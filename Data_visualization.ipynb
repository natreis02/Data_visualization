3. Criar um gráfico de dispersão mostrando a relação entre 'height' e 'weight' usando Seaborn.

```python
import seaborn as sns

# Definindo a paleta de cores "husl"
palette = sns.color_palette("husl", 5)

# Criando o gráfico de dispersão com a paleta de cores personalizada
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='height', y='weight', hue='gender', data=df, palette=palette, s=100)

# Melhorando a aparência do gráfico
scatter_plot.set_title('Relação entre Altura e Peso por Gênero')
scatter_plot.set_xlabel('Altura (cm)')
scatter_plot.set_ylabel('Peso (kg)')
plt.legend(title='Gênero')
plt.show()
```

4. Criar um gráfico interativo de barras mostrando a contagem de indivíduos por
'gender' usando Plotly.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Contando o número de ocorrências de cada gênero
gender_counts = df['gender'].value_counts()

# Usando a paleta "husl" do seaborn
colors = sns.color_palette('husl', n_colors=len(gender_counts))

# Criando o gráfico de barras
plt.figure(figsize=(8, 6))
bars = plt.bar(gender_counts.index, gender_counts.values, color=colors)

# Adicionando rótulos e título
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.title('Contagem de Gêneros no Conjunto de Dados')


# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import seaborn as sns

# Contando o número de ocorrências de cada gênero
gender_counts = df['gender'].value_counts()

# Usando a paleta "husl" do seaborn para definir cores
colors = sns.color_palette('husl', n_colors=len(gender_counts)).as_hex()

# Criando o gráfico de barras interativo com Plotly
fig = go.Figure(data=[
    go.Bar(x=gender_counts.index, y=gender_counts.values, marker_color=colors)
])

# Adicionando layout e título
fig.update_layout(
    title='Contagem de Gêneros no Conjunto de Dados',
    xaxis_title='Gênero',
    yaxis_title='Contagem'
)

# Exibindo o gráfico interativo
fig.show()
```

3. Criar um gráfico de decomposição sazonal usando Seaborn.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Dados fictícios de vendas ao longo do tempo
np.random.seed(0)
tempo = np.arange(1, 13)
vendas = np.random.randint(100, 1000, size=12)

# Usando a paleta "husl"
colors = sns.color_palette('husl', n_colors=len(tempo)).as_hex()

# Criando o gráfico de linhas com Matplotlib
plt.figure(figsize=(10, 6))
for i in range(len(tempo)):
    plt.plot(tempo[i:i+2], vendas[i:i+2], marker='o', linestyle='-', color=colors[i])

# Adicionando título e rótulos aos eixos
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('Mês')
plt.ylabel('Vendas')

# Adicionando grade
plt.grid(True)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

4. Criar um gráfico interativo que permita a exploração dos dados de vendas ao longo do tempo usando Plotly.

```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import seaborn as sns

# Dados fictícios de vendas ao longo do tempo
np.random.seed(0)
tempo = np.arange(1, 13)
vendas = np.random.randint(100, 1000, size=12)

# Usando a paleta "husl" do seaborn para definir cores
colors = sns.color_palette('husl', n_colors=len(tempo)).as_hex()

# Criando o gráfico interativo com Plotly
fig = go.Figure()

# Adicionando linhas de vendas ao longo do tempo
fig.add_trace(go.Scatter(x=tempo, y=vendas, mode='lines+markers', marker=dict(color=colors), name='Vendas'))

# Adicionando layout e título
fig.update_layout(
    title='Vendas ao Longo do Tempo',
    xaxis_title='Mês',
    yaxis_title='Vendas',
    hovermode='closest'
)

# Exibindo o gráfico interativo
fig.show()
```
