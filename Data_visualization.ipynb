```python
import seaborn as sns

# Definindo a paleta de cores "husl"
palette = sns.color_palette("husl", 5)

# Criando o gr√°fico de dispers√£o com a paleta de cores personalizada
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='height', y='weight', hue='gender', data=df, palette=palette, s=100)

# Melhorando a apar√™ncia do gr√°fico
scatter_plot.set_title('Rela√ß√£o entre Altura e Peso por G√™nero')
scatter_plot.set_xlabel('Altura (cm)')
scatter_plot.set_ylabel('Peso (kg)')
plt.legend(title='G√™nero')
plt.show()
```

3. Criar um gr√°fico de decomposi√ß√£o sazonal usando Seaborn.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Dados fict√≠cios de vendas ao longo do tempo
np.random.seed(0)
tempo = np.arange(1, 13)
vendas = np.random.randint(100, 1000, size=12)

# Usando a paleta "husl"
colors = sns.color_palette('husl', n_colors=len(tempo)).as_hex()

# Criando o gr√°fico de linhas com Matplotlib
plt.figure(figsize=(10, 6))
for i in range(len(tempo)):
    plt.plot(tempo[i:i+2], vendas[i:i+2], marker='o', linestyle='-', color=colors[i])

# Adicionando t√≠tulo e r√≥tulos aos eixos
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('M√™s')
plt.ylabel('Vendas')

# Adicionando grade
plt.grid(True)

# Exibindo o gr√°fico
plt.tight_layout()
plt.show()
```

4. Criar um gr√°fico interativo que permita a explora√ß√£o dos dados de vendas ao longo do tempo usando Plotly.

```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import seaborn as sns

# Dados fict√≠cios de vendas ao longo do tempo
np.random.seed(0)
tempo = np.arange(1, 13)
vendas = np.random.randint(100, 1000, size=12)

# Usando a paleta "husl" do seaborn para definir cores
colors = sns.color_palette('husl', n_colors=len(tempo)).as_hex()

# Criando o gr√°fico interativo com Plotly
fig = go.Figure()

# Adicionando linhas de vendas ao longo do tempo
fig.add_trace(go.Scatter(x=tempo, y=vendas, mode='lines+markers', marker=dict(color=colors), name='Vendas'))

# Adicionando layout e t√≠tulo
fig.update_layout(
    title='Vendas ao Longo do Tempo',
    xaxis_title='M√™s',
    yaxis_title='Vendas',
    hovermode='closest'
)

# Exibindo o gr√°fico interativo
fig.show()
```
